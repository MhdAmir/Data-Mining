# Let's Code

```python
import pandas as pd

dataset = pd.read_csv('titanic.csv')

dataset
```

```python
data = pd.DataFrame(dataset, columns=['Name', 'Sex', 'Age', 'Pclass', 'Fare'])
print(data)
```

```python
Class = pd.DataFrame(dataset, columns=['Survived'])
print(Class)
```

```python
dataset['Total'] = dataset['SibSp'] + dataset['Parch']
total_sum = dataset['Total'].sum()
print(f"Jumlah akumulasi dari kolom SibSp dan Parch: {total_sum}")
```

```python
Pclass = pd.DataFrame(dataset, columns=['Pclass'])
high = Pclass.loc[Pclass['Pclass'] == 1].count()
middle = Pclass.loc[Pclass['Pclass'] == 2].count()
low = Pclass.loc[Pclass['Pclass'] == 3].count()
print(high)
print(middle)
print(low)
```

```python
Sex = pd.DataFrame(dataset, columns=['Sex'])
male = Sex.loc[Sex['Sex'] == 'male'].count()
female = Sex.loc[Sex['Sex'] == 'female'].count()
print(male)
print(female)
```

```python
Passenger = pd.DataFrame(dataset, columns=['Survived'])
survived = Passenger.loc[Passenger['Survived'] == 1].count()
not_survived = Passenger.loc[Passenger['Survived'] == 0].count()

print(survived)
print(not_survived)
```

```python
dataset.plot(x='PassengerId', y = "Sex", kind='scatter', c='Survived', colormap='Paired')
```

```python
dataset.plot(x='PassengerId', y='Sex', kind='scatter', c='Pclass', colormap='viridis')
```

```python
dataset.dropna(subset=['Age']).plot(x='PassengerId', y='Age', kind='scatter', c='Pclass', colormap='viridis')
```
