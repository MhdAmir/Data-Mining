```python
import pandas as pd

dataset = pd.read_csv('titanic.csv')

dataset
```

```python
rows, cols = dataset.shape
print(f"Rows: {rows}, Columns: {cols}")
```

```python
data = pd.DataFrame(dataset, columns=['Age', 'Fare'])
data
```

```python
Class = pd.DataFrame(dataset, columns=['Survived'])
Class
```

```python
mean_age_by_class = dataset.groupby('Survived')['Age'].transform('mean')
data['Age'].fillna(mean_age_by_class, inplace=True)
data
```

```python
mean_age_by_class = dataset.groupby('Pclass')['Age'].transform('mean')
data['Age'].fillna(mean_age_by_class, inplace=True)
data
```

```python
data_min_max = (data - data.min()) / (data.max() - data.min())
data_min_max
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

min_max_with_library = MinMaxScaler()
data_min_max_lib = pd.DataFrame(min_max_with_library.fit_transform(data), columns=['Age', 'Fare'])
print('Min-Max Scaling With Library')
data_min_max_lib
```

```python
ZScoreReal = (data-data.mean())/data.std()
ZScoreReal
```

```python
standard_scaler = StandardScaler()
data_zscore_with_library = pd.DataFrame(standard_scaler.fit_transform(data), columns=['Age', 'Fare'])
print("\nZ-Score Scaling:")
data_zscore_with_library
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

data_sigmoidal = data.apply(sigmoid)
print("\nSigmoidal Scaling:")
data_sigmoidal
```

```python
import numpy as np

mean = data.mean()
std = data.std()

x = (data - mean) / std

newdata = (1 - np.exp(-x)) / (1 + np.exp(-x))

print("\nSigmoidal Scaling with Formula:")
newdata
```
